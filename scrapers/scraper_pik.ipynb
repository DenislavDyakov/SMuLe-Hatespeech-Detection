```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = "https://pik.bg/"
response = requests.get(base_site)
response
```

```python

response.text
```

```python
response.content
```

```python
str(response.content, 'utf-8', 'ignore')
```

```python
soup = BeautifulSoup(str(response.content, 'utf-8', 'ignore'))
```

```python
type(soup)
```

```python
soup.find_all(class_ = 'title')
```

```python
links = soup.find_all('a', class_ = 'box')
```

```python
link_list = [(l['href'] , l['title']) for l in links if 'news' in l['href'] ]
```

```python
link_list = set(link_list)
```

```python
link_list = list(link_list)
```

```python
link_list
```

# Multiple page scraping

```python
pik_results = []
```

```python
for record in link_list:
    url = record[0]
    print(url)
    title = record[1]
    print(title)
    print('-----------------')
    details = requests.get(url)
    details_html = details.content
    details_soup = BeautifulSoup(details_html, 'lxml')
    details_pars = details_soup.find_all("p")
    text = [p.text for p in details_pars if len(p.text) >4]
    print(text)
```
